Задание 1.
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
# исхожу из формата aXXXaa XXXrus
# их больше судя по вики, но все обрабатывать - времени жалко
def check_number(text):
    pattern = r'([а-яёА-ЯЁ]{1})(\d{3})([а-яёА-ЯЁ]{2})\s(\d{1,3})\s?(rus|RUS)'
    result=re.match(pattern, text)
    if result:
        return [result.group(2),result.group(4)]
    else:
        return None

print(check_number('dм288ах 25rus'))
print(check_number('sdfsdfsd'))
print(check_number('м288ах 125 sdffsdfsdfsdf'))
print(check_number('м2888ах 25rus'))
print(check_number('м288ах 25rus'))
print(check_number('м288ах 125 rus'))
```

Задание 2.
Постарайтесь написать (в крайнем случае найти) самый полный шаблон для проверки e-mail адресов через регулярное выражение, которая проверяет e-mail адрес на валидность. Можно руководствоваться следующими минимальными требованиями:
• адрес должен содержать специальный символ “@”, который отделяет имя пользователя почтовой системы от доменного имени;
• адрес не должен содержать следующие символы: “пробел”, “,”, “:”, “;”, “!”, “#”, “%”, “*”, “(”, “)”, “=”, “+”, “{”, “}”, " " ", " ’ ", " / ", " \ " и “|”;
• адрес должен состоять только из латинских символов;
• после символа “@” должна быть как минимум одна “.”;
• после последней точки должно быть не менее 2-х и не более 4-х символов, причем наличие цифр не допускается;
• между последней точкой и символом “@” должно быть не менее 2-х символов
• слева от “@” должно быть не менее четырех символов

```python
# сделал почти всё сам (чтобы не сильно усложнять), вроде потестил - работает
email_pattern=r'^(([a-zA-Z]+)[\.\-]?){4,}\@((([a-zA-Z\-0-9]{2,}\.)+[a-zA-Z]{2,4}))$'

print(re.match(email_pattern, "test@rambler.ru"))
print(re.match(email_pattern, "test@rur.ru1"))
```

**Задание 3. **
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
# сделал case insensitive, предполагается, что цифры - тоже слова (легко изменить, если нужны только буквенные слова)
tst = 'какой-то очень очень длинный длинный 123 12314 12314 текст для Для для о о проверки текст текстовый'
def remove_dupl(text):
    return re.sub(r'(?i)\b(\w+)\b(?:\s+\1\b)+', r'\1', text)

print(tst,'\n',remove_dupl(tst), sep='')
```

Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:
+7(xxx)-xxx-xx-xx

Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры валидных форматов:

• +7 955 555-55-55;
• 9555555555;
• 8(955)555-55-55;
• +7 955 555 55 55;
• 7(955) 555-55-55;
• +7 955+555+55+55.

```python
# учел .+-() и \s
# на большинство вариантов ввода должно работать

check_text = ['+7 955+555-55-55', '9555555556','8(955)555-55-57','+7 955 555 55 58','7(955) 555-55-59','+7 955+555+55+55', '+7 955b555+55+10', '+7 95 555+55+11', '7955555-5559']

def check_def(text):
    pattern = r'^(?:\+7|7|8){0,2}[\s\.\+\-\(\)]{0,2}(\d{3})[\s\.\+\-\(\)]{0,2}(\d{3})[\s\.\+\-\(\)]?(\d{2})[\s\.\+\-\(\)]?(\d{2})'
    return re.sub(pattern, r'+7(\1)-\2-\3-\4', text) if re.match(pattern,text) else None

for i, v in enumerate(check_text):
    res = check_def(check_text[i])
    print(f'{v}: \t {res if res else "Некорректный номер"}')
```
