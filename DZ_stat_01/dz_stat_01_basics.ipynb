Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
q1 = df_flt['Rectal temperature'].quantile(0.25)
q3 = df_flt['Rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
df_flt[~df_flt['Rectal temperature'].between(lower_bound, upper_bound, inclusive=True) & df_flt['Rectal temperature'].notna()]
```

```python
# Не очень разбираясь в лошадях, предположу, что (за исключением пропущенных значений) выбросы температуры вполне легитимны.
# Они не настолько сильно отличаются от среднего, чтобы вызывать подозрение.
```

```python
q1 = df_flt['Pulse'].quantile(0.25)
q3 = df_flt['Pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
df_flt[~df_flt['Pulse'].between(lower_bound, upper_bound, inclusive=True) & df_flt['Pulse'].notna()]
```

Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df_flt.describe()
```

```python
# В целом, из выбраного дата-сета (отстраненно) интересны Surgery, Age, Outcome, Surgical lesion.
# Они же являются наиболее полными данными.
# Остальные отсутствующие значения попробуем заполнить медианой с максимальной разбивкой по группам, указанным выше.

# Итого:
# Surgery - 1 пропуск, можно, конечно, заполнить значением из Surgical lesion, но лучше удалить (поскольку это наш основной столбец)
# Age - есть все значения.
# Rectal temperature - нет 60/300 значений. Довольно много. Попробуем заполнить медианным по группам с учетом Temperature of extremities, возрасту, хирург. вмеш, outcome
# Pulse - нет 24/300 значений. Заполним медианным по surgery, age, outcome.
# Temperature of extremities - нет 56/300. Заполним медианным по surgery, age, outcome (плюс проверим рект. температуру)
# Pain - нет 55/300. Оценочная характеристика. Удалить бы, конечно, но слишком много пропушенных будет. Заполним медианным по surgery, age, outcome (плюс проверим рект. температуру)
# Abdominal distension -аналогично pain
# Outcome - есть все, кроме одного. Эту запись удалим - непонятно, какой был итог осмотра, а 1 из 300 не так много.
# Surgical lesion - есть все.
for col in df_flt.columns:
    pct_missing = df_flt[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
df_res=df_flt
df_flt=df_flt.dropna(subset=['Surgery', 'Outcome', 'Surgical lesion'])
df_res['Rectal temperature'].fillna(df_flt.groupby(['Surgery','Age','Outcome','Surgical lesion'])['Rectal temperature'].transform('median'), inplace=True)
df_res['Pulse'].fillna(df_flt.groupby(['Surgery','Age','Outcome','Surgical lesion'])['Pulse'].transform('median'), inplace=True)
df_res['Temperature of extremities'].fillna(df_flt.round({'Rectal temperature': 0}).groupby(['Surgery','Age','Outcome','Surgical lesion','Rectal temperature'])['Temperature of extremities'].transform('median'), inplace=True)
df_res['Pain'].fillna(df_flt.groupby(['Surgery','Age','Outcome','Surgical lesion'])['Pain'].transform('median'), inplace=True)
df_res['Abdominal distension'].fillna(df_flt.groupby(['Surgery','Age','Outcome','Surgical lesion'])['Abdominal distension'].transform('median'), inplace=True)
# добиваем остатки за счет отстутвующих групп
df_res['Rectal temperature'].fillna(df_flt.groupby(['Surgery','Age'])['Rectal temperature'].transform('median'), inplace=True)
df_res['Pulse'].fillna(df_flt.groupby(['Surgery','Age'])['Pulse'].transform('median'), inplace=True)
df_res['Temperature of extremities'].fillna(df_flt.groupby(['Surgery','Age'])['Temperature of extremities'].transform('median'), inplace=True)
df_res['Pain'].fillna(df_flt.groupby(['Surgery','Age'])['Pain'].transform('median'), inplace=True)
df_res['Abdominal distension'].fillna(df_flt.groupby(['Surgery','Age'])['Abdominal distension'].transform('median'), inplace=True)

df_res
```
